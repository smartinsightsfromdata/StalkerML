```python
import sqlite3
from os import path

from html2corpus import HTML2Corpus
from corputil import StreamCorpus

db_file = path.join('..', 'Crawler', 'data', 'GermanNews.sqlite')
conn = sqlite3.connect(db_file)
cursor = conn.cursor()
```

```python
def get_data(start, end):
    query = 'SELECT title, html FROM Articles LEFT JOIN Sources ON Articles.source = Sources.id WHERE date between ? and ?'
    cursor.execute(query, (start, end))

    count = 0
    data = cursor.fetchmany(100)

    while len(data) > 0:
        for row in data:
            title, html = row
            yield html
        data = cursor.fetchmany(100)

corpus = StreamCorpus(HTML2Corpus(get_data('2015-10-25', '2015-10-27')))
for i, doc in enumerate(corpus):
    if i % 100 is 0:
        print(doc + '\n')
        
conn.close()
```
